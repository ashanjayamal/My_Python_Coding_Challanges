# Highly Divisible Triangular Number
### Problem 12
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number
would be $1 + 2 + 3 + 4 + 5 + 6 + 7 28$. The first ten terms would be:
$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55$$
Let us list the factors of the first seven triangle numbers:
 1: 1, 3
 3: 1, 3 
10: 1, 2, 5, 10
15: 1, 3, 5, 15
21: 1, 3, 7, 21
28:1 24 7 14,28
We can see that 28 is the first triangle number to have over five divisors.
What is the value of the first triangle number to have over five hundred divisors?

```python
def no_of_divisors(n):
    count=0
    for i in range(1,n):
        if n%i==0:
            count+=1
    return count

def Highly_Divisible_Triangular_Number(upperlimit):
    triagle_number=0
    i=1
    while True:
        triagle_number+=i
        noOfDivisors=no_of_divisors(triagle_number)
        print(triagle_number,noOfDivisors)
        if noOfDivisors>=upperlimit:
            print(i,"th triagle number is",triagle_number,".It has",noOfDivisors,"divisors")
            return triagle_number
        i+=1

```

```python
Highly_Divisible_Triangular_Number(5)
```
