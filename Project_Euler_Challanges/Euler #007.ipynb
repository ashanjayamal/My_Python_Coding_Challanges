# 10001st Prime
### Problem 7
By listing the first six prime numbers: $2, 3, 5, 7, 11,$ and $13,$ we can see that the 6th prime is $13$.
What is the $10001$ st prime number?

```python
from cmath import sqrt


def Isprime(n):
    if n==2 or n==3 or n==5: 
        return True
    elif n%2==0 or n%3==0 or n%5==0: # Many numbers are divided by these three numbers
        return False
    maxfactor=abs(sqrt(n))
    i=2
    while i<=maxfactor:

        if n%i==0:
            return False
            
        
        i+=1

    return True
    
def nth_Prime(upperlimit):
    j=0 #jth prime
    k=2 # check k is prime or not
    while j<upperlimit:
        if Isprime(k)==True:
            j+=1
            print(j,"th prime is",k)
        k+=1

    print("The",upperlimit,"th prime is",k-1)
    return(k-1)
```

```python
nth_Prime(6)
```

```python
nth_Prime(1001)
```
